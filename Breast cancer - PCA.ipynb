```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
breast = load_breast_cancer()
breast_data=breast.data
breast_label=breast.target
```

## 1. Process data

```python
breast_data
```

```python
breast_label.shape
```

```python
labels = np.reshape(breast_label,(569,1))
```

```python
labels
```

```python
final_breast_data = np.concatenate([breast_data, labels], axis = 1)
```

```python
final_breast_data
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
breast_dataset
```

```python
features = breast.feature_names
```

```python
features.shape
```

```python
features_labels = np.append(features, 'label')
features_labels
```

```python
b = features_labels
```

```python
breast_dataset['label'].replace(0, 'Benign', inplace = True)
breast_dataset['label'].replace(1, 'Malignant', inplace = True)
```

```python
breast_dataset
```

```python
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
feat_cols
```

```python
normalised_breast = pd.DataFrame(x, columns = feat_cols)
normalised_breast
```

## 2. PCA 

```python
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principalComponents_breast
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast, columns = ['PC1','PC2'])
principal_breast_Df
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1', fontsize=20)
plt.ylabel('Principal Component - 2', fontsize=20)
plt.title('Principal Component Analysis of Breast Cancer Dataset', fontsize = 20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets, colors):
    indicesToKeep = breast_dataset['label'] ==target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'PC1'], principal_breast_Df.loc[indicesToKeep, 'PC2'], c = color, s = 50) 
plt.legend(targets,prop={'size': 15})
```

```python
fig = plt.figure(figsize=(15, 5))
plt.imshow(pca_breast.components_, interpolation = 'none', cmap = 'plasma')


plt.gca().set_xticks(np.arange(-.5, len(features)));
plt.gca().set_yticks(np.arange(0.5, 2));
plt.gca().set_xticklabels(features_labels, rotation=90, ha='left', fontsize=12);
plt.gca().set_yticklabels(['PC1', 'PC2'], va='bottom', fontsize=12);

plt.colorbar(orientation='horizontal', ticks=[pca_breast.components_.min(), 0,
                                             pca_breast.components_.max()], pad=0.65);
```
